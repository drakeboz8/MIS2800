```python
#Block 1
# Instructions: There is a typo in the table.  Hint: a word is missing.
%%sql
DROP TABLE IF EXISTS Marketing_Brands;
CREATE TABLE Marketing_Brands (
  brand_id INTEGER PRIMARY KEY,
  brand_name TEXT,
  brand_type TEXT,
  brand_budget INTEGER
) STRICT;
```

```python
#Block 2
# Instructions: Find and fix the error.

%%sql
INSERT OR REPLACE INTO Marketing_Brands (brand_id, brand_name, brand_type, brand_budget)
VALUES
  (1, 'Doritos', 'Snacks', 7000000),
  (2, 'AWS', 'Technology', 15000000),
  (3, 'Pepsi', 'Beverages', 2000000),
  (4, 'He_Gets_Us', 'Religious', 1000000),
  (5, 'Dr_Pepper', 'Beverages', 5000000),
  (6, 'IBM', 'Technology', 100000),
  (7, 'Google', 'Technology', 8000000);
```

```python
#Block 3
# -- Instructions: Fix the code to select all the columns
# from the Marketing_Brands table
%%sql
#Block 4
# -- Instructions: Fix the code to display only the brand names and
# brand types that had budget over 700000SELECT * FROM Marketing_Brands;
```

```python
#Block 4
# -- Instructions: Fix the code to display only the brand names and
# brand types that had budget over 700000

%%sql
SELECT brand_name, brand_type FROM Marketing_Brands WHERE brand_budget > 700000;
```

```python
#Block 5
%%sql
-- Instructions: Fix the code to find the brand types that are Religious
SELECT brand_name, brand_type FROM Marketing_Brands WHERE brand_type = 'Religious';
```

```python
#Block 6
# -- Instructions: Fix the code to display the
# brand names that have a budget under 1000000
%%sql
SELECT FROM Marketing_Brands WHERE brand_budget < 1000000;
```

```python
#Block 7
# Instructins: Fix the code so that this will run SQLite3 in
# Colab without having to wrap in Python.  Hint: It's Magic!
%%sql
SELECT * FROM Marketing_Brands WHERE brand_budget > 7000000
```

```python
#Block 8
# Instructions: Find the brands where the brand type is
# Snacks or Beverages.  Display only the brand names.
%%sql
SELECT brand_name FROM Marketing_Brands WHERE brand_type = 'Snacks' OR brand_type = 'Beverages';
```

```python
#Block 9
# Instructions: Sort the table by brand budget in
# DESC order. Display only the brand_name and brand_budget.
%%sql
SELECT brand_name, brand_budget FROM Marketing_Brands ORDER BY brand_budget DESC;
```

```python
#Block 10
# Using SQLite3 (and all its magic) to create a
# table showing all the columns in ascending order by budget.
%%sql
SELECT * FROM Marketing_Brands ORDER BY brand_budget ASC;
```
